```python
ds=pd.DataFrame({'a1':[68,70,'A','B',88], 'a2':[98,'A','A','A',85], 'a3':['A',70,'A',66,85],

'a4':[62,74,'A',55,78]})

ds
```

```python
ds=ds.apply(pd.to_numeric,errors='coerce')
ds
```

```python
ds.isnull().sum()
```

```python
print("med ",ds["a1"].median())
ds['a1'].fillna(ds['a1'].median(),inplace=True)
ds
```

```python
print("mean ",ds["a2"].mean())
ds['a2'].fillna(ds['a2'].mean(),inplace=True)
ds
```

```python
print("mode ",ds["a5"].mode())
ds['a5'].fillna(ds['a5'].mode()[0],inplace=True)
ds
```
